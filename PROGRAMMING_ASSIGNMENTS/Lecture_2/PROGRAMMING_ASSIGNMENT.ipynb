```python
import pandas as pd
import matplotlib.pyplot as plt
dir = "../tic-tac-toe.data"
df =  pd.read_csv(dir, sep=",")

corr_matrix = df.corr()


df.describe()
```

```python
# import module
from tabulate import tabulate
 
# assign data
mydata = [
    [corr_matrix["win"].a, corr_matrix["win"].b, corr_matrix["win"].c], 
    [corr_matrix["win"].d, corr_matrix["win"].e, corr_matrix["win"].f], 
    [corr_matrix["win"].g, corr_matrix["win"].h, corr_matrix["win"].i]
]
 
# display table
print("chances of winning based on placement")
print(tabulate(mydata, tablefmt="grid"))
```
