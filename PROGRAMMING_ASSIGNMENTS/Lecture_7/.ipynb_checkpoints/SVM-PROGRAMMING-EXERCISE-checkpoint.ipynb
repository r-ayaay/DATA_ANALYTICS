**TASK: Take 10 moderately sized subsamples your dataset and create a correlation plot for each subsample**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("obesity.csv")
df = pd.get_dummies(df, columns=['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'])
target = 'NObeyesdad'
num_cols = df.select_dtypes(include='number').columns
num_cols = [col for col in num_cols if col != target]

subsample_size = len(df) // 10
subsamples = [df.sample(n=subsample_size, replace=False) for _ in range(10)]

for i, subsample in enumerate(subsamples):
    correlation_matrix = subsample[num_cols].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f"Correlation Plot - Subsample {i+1}")
    plt.show()
```

**TASK (Regression): If your problem is a regression problem visually the distribution of your target variabe**

```python
plt.figure(figsize=(8, 6))
df['NObeyesdad'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Target Variable')
plt.xlabel('Classes')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

**TASK: Take 10 moderately sized subsamples and create a pairplot of the feature variable in relation to the target variable**

```python
subsamples = [df.sample(n=subsample_size, replace=False) for _ in range(10)]

for i, subsample in enumerate(subsamples):
    plt.figure(figsize=(12, 8))
    sns.pairplot(subsample, hue='NObeyesdad', palette='Set1')
    plt.title(f"Pairplot - Subsample {i+1}")
    plt.show()
```

**TASK: Create a clustermap with seaborn to explore the relationships between variables.**

```python
df = pd.get_dummies(df, columns=['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS'])
target = 'NObeyesdad'
num_cols = df.select_dtypes(include='number').columns
num_cols = [col for col in num_cols if col != target]
correlation_matrix = df[num_cols].corr()
plt.figure(figsize=(12, 10))
sns.clustermap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Clustermap of Variable Relationships')
plt.show()
```
